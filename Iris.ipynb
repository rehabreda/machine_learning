```python
from sklearn.linear_model import  LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import svm
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
```

```python
 
sns.heatmap(data.corr(),annot=True,cmap='cubehelix_r')
plt.show()
```

```python
# split data 
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.3)

```

```python
#svm
model=svm.SVC()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print('the accuracy of svm is ',metrics.accuracy_score(y_test,pred))
```

```python
#logidtic regression
model=LogisticRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print('the accuracy of logistic regression is ',metrics.accuracy_score(y_test,pred))
```

```python
#decision tree 
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print('the accuracy of decision tree is ',metrics.accuracy_score(y_test,pred))
```

```python
# kneighbors
model=KNeighborsClassifier()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print('the accuracy of Kneighbors is ',metrics.accuracy_score(y_test,pred))
```

```python
acc=[]
for i in list(range(1,11)):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    acc.append(metrics.accuracy_score(y_test,model.predict(x_test)))
    
    
    
plt.plot(range(1,11),acc) 

    
    
```
